```python
#Importações
import pandas as pd
from sklearn.model_selection import train_test_split
```

### As bases de dados corporas são:
1. https://github.com/roneysco/Fake.br-Corpus
2. https://github.com/jghm-f/FACTCK.BR/
3. https://paperswithcode.com/dataset/x-fact


```python
#https://github.com/roneysco/Fake.br-Corpus
data1 = pd.read_table('https://raw.githubusercontent.com/roneysco/Fake.br-Corpus/master/preprocessed/pre-processed.csv', sep=',')

#https://github.com/jghm-f/FACTCK.BR/
data2 = pd.read_table('https://raw.githubusercontent.com/jghm-f/FACTCK.BR/master/FACTCKBR.tsv')

#https://paperswithcode.com/dataset/x-fact
data3 = pd.read_table('https://raw.githubusercontent.com/utahnlp/x-fact/main/data/x-fact/dev.all.tsv', sep='\t', error_bad_lines=False)
```

###Tratamento dos dados do data1:

Considerando apenas 2 atributos [noticias e classe]. Onde as noticías são os corpus de texto e o atributo alvo será a classe falso como valor 0 ou verdadeiro como valor 1.

```python
data1 = data1.drop(['index'], axis=1)
data1.columns = ['classe', 'noticias']
data1 = data1.reindex(columns=['noticias', 'classe'])
data1['classe'] = data1['classe'].map({'fake': 0, 'true': 1})
data1
```

```python
#Certificamos que existem apenas 2 objetos
data1['classe'].unique()
```

```python
#Contando quantos falsos e quantos verdadeiros o dataset possui
contagem = data1['classe'].value_counts()
print(contagem)
```

```python
#Contando quantos falsos e quantos verdadeiros o dataset possui
#amostra = data1.sample(n=5, random_state=42)
amostra = data1.sample(n=15)
amostra
```

###Tratamento dos dados do data2:

Considerando apenas 2 atributos [noticias e classe]. Onde as noticías são os corpus de texto e o atributo alvo será a classe falso como valor 0 ou verdadeiro como valor 1.

```python
data2 = data2.drop(['URL', 'Author', 'datePublished', 'title', 'reviewBody', 'ratingValue', 'bestRating'], axis=1)
data2 = data2.reindex(columns=['claimReviewed', 'alternativeName'])
data2.columns = ['noticias', 'classe']
data2 = data2[(data2['classe'] == 'verdadeiro') | (data2['classe'] == 'Verdadeiro') | (data2['classe'] == 'falso') | (data2['classe'] == 'Falso')] #Verificando com o unique percebemos todos estes objetos
data2['classe'] = data2['classe'].map({'verdadeiro': 1, 'Verdadeiro': 1, 'Falso': 0, 'falso': 0})
data2['noticias'] = data2['noticias'].str.replace('#Verificamos:', '') #Limpeza prévia nos corpus
data2['noticias'] = data2['noticias'].str.replace('#VerificamosENEM:', '') #Limpeza prévia nos corpus
data2
```

```python
#Certificamos que existem apenas 2 objetos
data2['classe'].unique()
```

```python
#Contando quantos falsos e quantos verdadeiros o dataset possui
contagem = data2['classe'].value_counts()
print(contagem)
```

```python
#Verificando uma pequena amostra do dataframe
#amostra = data2.sample(n=20, random_state=42)
amostra = data2.sample(n=15)
amostra
```

###Tratamento dos dados do data3:

Considerando apenas 2 atributos [noticias e classe]. Onde as noticías são os corpus de texto e o atributo alvo será a classe falso como valor 0 ou verdadeiro como valor 1.

```python
data3 = data3[(data3['language'] == 'pt')]
data3 = data3[(data3['label'] == 'false') | (data3['label'] == 'true')] #Estamos considerando apenas os que são realmente verdadeiros e os que são classificados como realmente falsos.
data3 = data3.drop(['language', 'site', 'link_1', 'link_2', 'link_3', 'link_4', 'link_5', 'claimDate', 'reviewDate', 'claimant', 'claim'], axis=1)
```

```python
df = pd.DataFrame(data3)
df_novo = pd.DataFrame()
for index, row in df.iterrows():
    for i in range(1, 6):
        evidence_value = row[f"evidence_{i}"]
        label = row['label']
        df_temp = pd.DataFrame({'evidence': [evidence_value], 'label': [label]})
        df_novo = pd.concat([df_novo, df_temp], ignore_index=True)

data3 = df_novo
data3.columns = ['noticias', 'classe']
data3['classe'] = data3['classe'].map({'true': 1, 'false': 0})
data3
```

```python
#Certificamos que existem apenas 2 objetos
data3['classe'].unique()
```

```python
#Contando quantos falsos e quantos verdadeiros o dataset possui
contagem = data3['classe'].value_counts()
print(contagem)
```

###Unindo os 3 dataframes em um apenas

```python
dados_noticias = pd.concat([data1, data2, data3])
dados_noticias
```

```python
#Certificamos que existem apenas 2 objetos
dados_noticias['classe'].unique()
```

```python
#Contando quantos falsos e quantos verdadeiros o dataset possui
contagem = dados_noticias['classe'].value_counts()
print(contagem)
```

###Separando os dados em treino, validação e teste.

```python
X = dados_noticias["noticias"]
y = dados_noticias["classe"]

# Etapa 1: Separar 90%-10% para teste final
X_train_val, X_test, y_train_val, y_test = train_test_split(X, y, test_size=0.1, random_state=42, stratify=y)

# Etapa 2: Separar 90% restantes em 80%-20% para treinamento e validação
X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val, test_size=0.2, random_state=42, stratify=y_train_val)

# Os dados de treinamento estão em X_train e y_train
# Os dados de validação estão em X_val e y_val
# Os dados de teste final estão em X_test e y_test
```
